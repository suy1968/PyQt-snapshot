```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression,Ridge, Lasso
from mlxtend.evaluate import bias_variance_decomp
from sklearn.model_selection import train_test_split
```

```python
x = np.arange(1,11)
```

## Linear Regression

```python
plt.figure(figsize=(15,10))
i=1
# loop will run for different polynomial degree
for deg in [2,3,7,11]:    
    y = np.array([i**deg for i in range(10)])

    # Linear Model
    linear = LinearRegression().fit(x.reshape(-1,1),y)
    plt.subplot(2,2,i)
    plt.scatter(x,y)
    plt.plot(x, linear.predict(x.reshape(-1,1)) ,color='gray', linewidth=3)
    
    # train and test splitting
    xtrain, xtest, ytrain, ytest = train_test_split(x.reshape(-1,1),y)

    # calculating bias , variance and error
    e,b,v = bias_variance_decomp(linear, 
                         xtrain,
                         ytrain,
                         xtest,
                         ytest ,
                         loss='mse')
   
    plt.title("( Bias - "+str(b)+" )\n( Variance - "+str(v)+" )\n( Degree - "+str(deg)+" )")
    plt.xlabel('X',fontsize=12)
    plt.ylabel('Y',fontsize=12)
    plt.legend(['Best-Fit line','Points'],fontsize=14)
    
    plt.tight_layout()
    i+=1
    
plt.savefig("/home/suyash/Desktop/Linear.png",bbox_inches='tight')
```
